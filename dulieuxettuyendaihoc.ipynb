**Tải dữ liệu xét tuyển đại hoc**

```
df = pd.read_csv('dulieuxettuyendaihoc.csv', index_col='STT')
df.head(10)
```

```
df.tail(10)
```

```
#Thống kê dữ liệu thiếu cho cột dân tộc và hiệu chỉnh dữ liệu thiếu như sau: Mặc định thiếu
# thì điền giá trị 0.

df['DT'].fillna(0, inplace=True)
df
```

```
# Thống kê dữ liệu thiếu cho biến T1 và hiệu chỉnh dữ liệu, lưu ý việc thay thế dữ liệu thiếu sử
#dụng phương pháp Mean
df['T1'].fillna(df['T1'].mean(), inplace=True)
df
```

```
# Hãy thực hiện xử lý lần lượt tất cả dữ liệu thiếu cho các biến về điểm số còn lại.
df[i]=df[i].fillna(df[i].mean())

df
```

```
#  Tạo các biến TBM1, TBM2, TBM3 tương ứng với trung bình môn của các năm lớp 10, 11 và
12.
df['TBM1']=(df['T1']*2+df['L1']+df['H1']+df['S1']+df['V1']*2+df['X1']+df['D1']+df['N1'])/10

df['TBM2']=(df['T2']*2+df['L2']+df['H2']+df['S2']+df['V2']*2+df['X2']+df['D2']+df['N2'])/10

df['TBM3']=(df['T3']*2+df['L3']+df['H3']+df['S3']+df['V3']*2+df['X3']+df['D3']+df['N3'])/10

df
```

```
##8. Tạo các biến xếp loại XL1, XL2 và XL3 dựa trên TBM1,TBM2 và TBM3 cho từng năm lớp
#10, 11, 12 như sau:
#• Nhỏ hơn 5.0 xếp loại: yếu (kí hiệu là Y)
#• Từ 5.0 đến dưới 6.5: trung bình (kí hiệu là TB)
#• Từ 6.5 đến dưới 8.0: khá (kí hiệu là K)
#• Từ 8.0 đến dưới 9.0: giỏi (kí hiệu là G)
#• Từ 9.0 trở lên: xuất sắc (kí hiệu là XS)
# Hàm xếp loại
def xeploai(diem):
    if diem < 5:
        return 'Y'
    elif diem < 6.5:
        return 'TB'
    elif diem < 8:
        return 'K'
    elif diem < 9:
        return 'G'
    else:
        return 'XS'
df['XL1'] = df['TBM1'].apply(xeploai)
df['XL2'] = df['TBM2'].apply(xeploai)
df['XL3'] = df['TBM3'].apply(xeploai)

df[['TBM1','XL1','TBM2','XL2','TBM3','XL3']].head(10)
```

```
## Tạo các biến US_TBM1, US_TBM2 và US_TBM3 để chuyển điểm trung bình các năm lớp
#10, 11 và 12 từ thang điểm 10 của Việt Nam sang thang điểm 4 của Mỹ. Sử dụng phương
#pháp Min-Max Normalization

df['US_TBM1'] = (df['TBM1'] - 0)/(10 - 0) * 4
df['US_TBM2'] = (df['TBM2'] - 0)/(10 - 0) * 4
df['US_TBM3'] = (df['TBM3'] - 0)/(10 - 0) * 4
df[['TBM1','US_TBM1','TBM2','US_TBM2','TBM3','US_TBM3']].head(10)
```

```
## 10. Tạo biến kết quả xét tuyển (kí hiệu là KQXT) nhằm xác định sinh viên đậu (giá trị “1”) và
# rớt ( giá trị “0”) vào các khối dựa trên điểm DH1, DH2 và DH3 như sau
# • Với khối A, A1 nếu [(DH1*2 + DH2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu,
# ngược lại là rớt
# • Với khối B nếu [(DH1 + DH2*2 + DH3)/4] lớn hơn hoặc bằng 5.0 thì đậu, ngược
# lại là rớt
# • Với khối khác nếu [(DH1+ DH2 + DH3)/3] lớn hơn hoặc bằng 5.0 thì đậu, ngược
# lại là rớt
# 1. Định nghĩa hàm xét tuyển dựa trên dữ liệu của từng sinh viên (từng dòng)
def xet_tuyen(row):
    if row['KT'] in ['A','A1']:
        diem = (row['DH1']*2 + row['DH2'] + row['DH3'])/4
    elif row['KT'] == 'B':
        diem = (row['DH1'] + row['DH2']*2 + row['DH3'])/4
    else:
        diem = (row['DH1'] + row['DH2'] + row['DH3'])/3

    return 1 if diem >= 5 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)
df[['KT','DH1','DH2','DH3','KQXT']].head()
```
