## Phần 4: Trực quan hóa dữ liệu nâng cao
1. Vẽ biểu đồ đường Simple cho biến T1
2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
d. Từ 8 trở lên = giỏi (ký hiệu “g”)
3. Lập bảng tần số cho biến phanloait1
4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1
5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

```
import matplotlib.pyplot as plt
bang = df['T1'].value_counts().sort_index()
plt.figure(figsize=(6,4))
plt.plot(bang.index, bang.values)
plt.xlabel('T1')
plt.ylabel('Count')
plt.title('Simple Line - T1')

plt.show()
```

```

def phan_loai_t1(diem):
    if diem < 5:
        return 'k'
    elif diem < 7:
        return 'tb'
    elif diem < 8:
        return 'kha'
    else:
        return 'g'
df['phanlopt1'] = df['T1'].apply(phan_loai_t1)
print(df[['T1', 'phanlopt1']])
```

```
bang_tanso = df['phanlopt1'].value_counts()

print("BẢNG TẦN SỐ BIẾN PHANLOPT1")
print(bang_tanso)
```

```
import pandas as pd
import matplotlib.pyplot as plt
bang = pd.crosstab(df['T1'], df['phanlopt1'])
bang.plot(
    kind='line',
    figsize=(8,5),
    marker='o'
)
plt.xlabel('T1')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```
nhom_k = df[df['phanlopt1'] == 'k']['T1'].value_counts().sort_index()
nhom_tb = df[df['phanlopt1'] == 'tb']['T1'].value_counts().sort_index()
nhom_kha = df[df['phanlopt1'] == 'kha']['T1'].value_counts().sort_index()
nhom_g = df[df['phanlopt1'] == 'g']['T1'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.stem(nhom_k.index, nhom_k.values, linefmt='g-', markerfmt='go', basefmt=' ')
plt.stem(nhom_tb.index, nhom_tb.values, linefmt='m-', markerfmt='mo', basefmt=' ')
plt.stem(nhom_kha.index, nhom_kha.values, linefmt='y-', markerfmt='yo', basefmt=' ')
plt.stem(nhom_g.index, nhom_g.values, linefmt='b-', markerfmt='bo', basefmt=' ')
plt.xlabel('T1')
plt.ylabel('Count')
plt.legend(['Kem', 'Trung binh', 'Kha', 'Gioi'])
plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
print("THỐNG KÊ MÔ TẢ T1")
print(df['T1'].describe())
print("\nMedian =", df['T1'].median())
print("Mode =")
print(df['T1'].mode())
print("\nVariance =", df['T1'].var())
print("Std =", df['T1'].std())
print("Range =", df['T1'].max() - df['T1'].min())
plt.figure(figsize=(8,4))
sns.boxplot(x=df['T1'])
plt.title('Boxplot của T1')
plt.show()
Q1 = df['T1'].quantile(0.25)
Q2 = df['T1'].quantile(0.50)
Q3 = df['T1'].quantile(0.75)
IQR = Q3 - Q1
Lower = Q1 - 1.5 * IQR
Upper = Q3 + 1.5 * IQR
print("Min =", df['T1'].min())
print("Q1 =", Q1)
print("Median =", Q2)
print("Q3 =", Q3)
print("Max =", df['T1'].max())
print("IQR =", IQR)
print("Lower Fence =", Lower)
print("Upper Fence =", Upper)
plt.figure(figsize=(8,4))
sns.histplot(df['T1'], bins=10, kde=True)
plt.title('Histogram của T1')
plt.show()
print("\nSkewness =", stats.skew(df['T1']))
print("Kurtosis =", stats.kurtosis(df['T1']))
plt.figure(figsize=(6,6))
stats.probplot(df['T1'], dist='norm', plot=plt)
plt.title('QQ Plot của T1')
plt.show()
```
